### Extracting months of the year, weekdays, morning, afternoon, evening and night.

```python
aq['Time'].sample(8)
```

```python
aq['Date']= pd.to_datetime(aq['Date'])
```

```python
aq['Time']= pd.to_datetime(aq['Time'])
```

```python
aq['year'] =pd.DatetimeIndex(aq['Date']).year
```

```python
aq['Month'] =pd.DatetimeIndex(aq['Date']).month
```

```python
aq['Month name'] =aq['Date'].dt.month_name(locale = 'English')
```

```python
aq['Days'] =pd.DatetimeIndex(aq['Date']).day
```

```python
aq['Weekday'] =pd.DatetimeIndex(aq['Date']).weekday_name
```

```python
aq.head()
```

#### getting morning, afternoon and night

```python
aq['Time no'] = (aq['Time'].dt.hour % 24 + 4) // 4  
aq['Time no'].replace({1: 'Late Night',
                      2: 'Early Morning',
                      3: 'Morning',
                      4: 'Noon',
                      5: 'Evening',
                      6: 'Night'}, inplace=True)
```
